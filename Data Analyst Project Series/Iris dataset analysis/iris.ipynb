```python
# scatterplot
colors = ['red', 'orange', 'blue']
species = ['virginica', 'versicolor', 'setosa']
```

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['sepal_length'], x['sepal_width'], c = colors[i], label=species[i])
    plt.xlabel('sepal_length')
    plt.ylabel('sepal_width')
    plt.legend()
```

OBSERVATIONS:
IN THE SCATTER PLOT, EACH POINT REPRESENTS AN INDIVIDUAL IRIS FLOWER WITH THE X-AXIS SHOWING THE SEPAL LENGTH AND THE Y-AXIS SHOWING THE SEPAL WIDTH. 
SETOSA: SETOSA IRIS FLOWERS USUALLY HAVE SHORTER SEPAL LENGTH AND WIDER SEPAL WIDTH COMPARED TO THE OTHER SPECIES. IN THE SCATTER PLOT, SETOSA FLOWERS          CLUSTER IN A SPECIFIC REGION WITH SHORTER SEPAL LENGTHS AND WIDER SEPAL WIDTHS.
VERSICOLOR: VERSICOLOR IRIS FLOWERS TYPICALLY HAVE AVERAGE SEPAL LENGTH AND SEPAL WIDTH. IN THE SCATTER PLOT, THEY ARE SCATTERED IN A MODERATE RANGEE               OF SEPAL LENGTHS AND SEPAL WIDTHS.
VIRGINICA: VIRGINICA IRIS FLOWERS TEND TO HAVE LONGER SEPAL LENGTH AND AVERAGE TO SLIGHTLY WIDER SEPAL WIDTH. IN THE SCATTER PLOT, VIRGINICA FLOWERS               ARE LOCATED IN A REGION WITH LONGER SEPAL LENGTHS AND SLIGHTLY WIDER SEPAL WIDTHS COMPARED TO OTHER SPECIES.

```python
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['petal_length'], x['petal_width'], c = colors[i], label=species[i])
    plt.xlabel('petal_length')
    plt.ylabel('petal_width')
    plt.legend()
```
